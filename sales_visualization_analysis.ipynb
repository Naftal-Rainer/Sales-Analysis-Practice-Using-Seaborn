# Introduction

To identify Which ship mode is best used by what product category and customer segment


```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
sales_data = pd.read_excel('sales.xlsx')
sales_data.head()
```

```python
sales_data.shape
```

Missing Values

```python
sales_data.isnull().sum()
```

```python
sales_data.fillna(sales_data['Product Base Margin'].mode(),inplace=True)
```

## Visualizations

### **Relationship between variables**

#### Sales vs Profit

```python
sns.scatterplot(x='Sales', y ='Profit', data=sales_data)
```

```python
sns.relplot(x='Sales', y ='Profit', data=sales_data)
```

```python
sns.relplot(x='Sales', y ='Profit', data=sales_data, hue='Order Priority')
```

```python
sns.relplot(x='Sales', y ='Profit', data=sales_data, hue='Order Priority',style='Order Priority')
```

Style using ship mode

```python
sns.relplot(x='Sales', y ='Profit', data=sales_data, hue='Order Priority',style='Ship Mode')
```

```python
sns.relplot(x='Sales', y ='Profit', data=sales_data, size='Discount')
```

```python
sns.relplot(x='Sales', y ='Profit', data=sales_data, size='Discount', sizes=(25,200))
```
